## Step 2: Variables and Calculation

This cell uses variables to perform a simple calculation.

**Important:** In Jupyter, later cells use the *current values stored in memory* (the kernel).  
If you change a value but **don’t re-run the cell**, the old value is still in memory.


````python
# Step 2: define inputs
hours_worked = 10
hourly_rate = 20

# Step 2: compute result used by later cells
total_pay = hours_worked * hourly_rate

print("Total pay:", total_pay)
````

````python
print("hours_worked =", hours_worked)
print("hourly_rate  =", hourly_rate)
print("total_pay    =", total_pay)
````

### Your task (shows why order matters)

1. **Edit** `hours_worked` in **Step 2**, but **do NOT run Step 2 yet**.
2. Run **Step 3** anyway. Notice: the message probably does **not** change.
3. Now run **Step 2**, then run **Step 3** again. Notice: the message updates.

This is the key idea: **changing text in a cell does nothing until you run the cell.**


## Step 3: Decision Logic (if / else)

This cell makes a decision based on `total_pay` (which was calculated in Step 2).

````python
# Step 3: decision based on the current value of total_pay in memory
if total_pay >= 200:
    print("Pay meets minimum target")
else:
    print("Pay does not meet minimum target")
````
